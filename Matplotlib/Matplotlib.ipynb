# Example: Draw a line in a diagram from position (1, 3) to position (8, 10):

```python
import matplotlib.pyplot as plt #
import numpy as np

# pass two arrays [1,3] and [8, 10] to the plot function.

xpoints = np.array([1,8])
ypoints = np.array([8,10])

plt.plot(xpoints, ypoints)# plot()function is used to draw points (markers) in a diagram.
# It has Parameter 1 is an array containing the points on the x-axis.

#Parameter 2 is an array containing the points on the y-axis.

plt.show()
```

 # Plotting Without Line

1. To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints,'o')
plt.show()
```

#**Matplotlib Marker**
#use the keyword argument(marker="o") marker to emphasize each point with a specified marker

```python
import matplotlib.pyplot as plt

xpoints=[1,3,5]
ypoints=[2,5,8]

plt.plot(xpoints,ypoints,marker="*")
plt.show()
```

#Format Strings fmt
#syntax: marker|line|color

```python
import matplotlib.pyplot as plt

xpoints=[1,3,5]
ypoints=[2,5,8]

plt.plot(xpoints,ypoints,"o:g")
plt.show()
```

#Marker Size :ms

```python
import matplotlib.pyplot as plt

xpoints=[1,3,5]
ypoints=[2,5,8]

plt.plot(xpoints,ypoints,marker='o',ms=10)
plt.show()
```

# #Marker Color
You can use the keyword argument **markeredgecolor** or the shorter **mec** to set the color of the edge of the markers:

```python
import matplotlib.pyplot as plt

xpoints=[1,3,5]
ypoints=[2,5,8]

plt.plot(xpoints,ypoints,marker='o',ms=10, mec='r')
plt.show()
```

#You can use the keyword argument **markerfacecolor** or the shorter **mfc** to set the color inside the edge of the markers

```python
import matplotlib.pyplot as plt

xpoints=[1,3,5]
ypoints=[2,5,8]

plt.plot(xpoints,ypoints,marker='o',ms=10, mec='r',mfc='y')
plt.show()
```

```python
import matplotlib.pyplot as plt

xpoints=[1,3,5]
ypoints=[2,5,8]

plt.plot(xpoints,ypoints,marker='*',ms=10, mec='r',mfc='hotpink')
plt.show()
```
