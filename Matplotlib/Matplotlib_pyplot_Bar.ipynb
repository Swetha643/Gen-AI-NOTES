#Creating Bars
With Pyplot, you can use the bar() function to draw bar graphs:

```python
import matplotlib.pyplot as plt

x = ["A", "B", "C", "D"]
y = [3, 8, 1, 10]

plt.bar(x,y)
plt.show()
```

```python
x = ["APPLES", "BANANAS"]
y = [400, 350]
plt.bar(x, y)
```

#Horizontal Bars -barh()
If you want the bars to be displayed horizontally instead of vertically, use the barh()

```python
import matplotlib.pyplot as plt

x = ["A", "B", "C", "D"]
y = [3, 8, 1, 10]

plt.barh(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["A", "B", "C", "D"]
y = [3, 8, 1, 10]

plt.bar(x,y,color="Hotpink")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ["A", "B", "C", "D"]
y = [3, 8, 1, 10]

plt.bar(x,y,color="Hotpink",width=0.1)
plt.show()
```

#Matplotlib Histograms
The hist() function will use an array of numbers to create a histogram, the array is sent into the function as an argument.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5]

# Create a histogram
plt.hist(data, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Sample Data')

# Show plot
plt.show()
```

**#Matplotlib Pie Charts**
#With Pyplot, you can use the pie() function to draw pie charts:

```python
import matplotlib.pyplot as plt

x = [35, 25, 25, 15]

plt.pie(x)
plt.show()
```

#Labels
Add labels to the pie chart with the labels parameter.

The labels parameter must be an array with one label for each wedge:

```python
import matplotlib.pyplot as plt

x = [35, 25, 25, 15]
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(x,labels=mylabels)
plt.show()
```

#Start Angle
As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter.

The startangle parameter is defined with an angle in degrees, default angle is 0:

```python
import matplotlib.pyplot as plt

x = [35, 25, 25, 15]
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(x,labels=mylabels,startangle=90)
plt.show()
```

#Explode
Maybe you want one of the wedges to stand out? The explode parameter allows you to do that.

The explode parameter, if specified, and not None, must be an array with one value for each wedge.

```python
import matplotlib.pyplot as plt

x = [35, 25, 25, 15]
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode=[0.2,0,0,0]

plt.pie(x,labels=mylabels,startangle=90,explode=myexplode)
plt.show()
```

#Shadow
Add a shadow to the pie chart by setting the shadows parameter to True:

```python
import matplotlib.pyplot as plt

x = [35, 25, 25, 15]
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode=[0.2,0,0,0]

plt.pie(x,labels=mylabels,startangle=90,explode=myexplode,shadow=True)
plt.show()
```

#Colors
You can set the color of each wedge with the colors parameter.

The colors parameter, if specified, must be an array with one value for each wedge:

```python
import matplotlib.pyplot as plt

x = [35, 25, 25, 15]
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode=[0.2,0,0,0]
mycolors=["hotpink","yellow","blue","green"]

plt.pie(x,labels=mylabels,startangle=90,explode=myexplode,shadow=True,colors=mycolors)
plt.show()
```

#Legend
To add a list of explanation for each wedge, use the legend() function:

```python
import matplotlib.pyplot as plt

x = [35, 25, 25, 15]
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode=[0.2,0,0,0]
mycolors=["hotpink","yellow","blue","green"]

plt.pie(x,labels=mylabels,startangle=90,explode=myexplode,shadow=True,colors=mycolors)
plt.legend()
plt.show()
```

#Legend With Header
To add a header to the legend, add the title parameter to the legend function.

```python
import matplotlib.pyplot as plt

x = [35, 25, 25, 15]
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode=[0.2,0,0,0]
mycolors=["hotpink","yellow","blue","green"]

plt.pie(x,labels=mylabels,explode=myexplode,shadow=True,colors=mycolors)
plt.legend(title="Four fruits")
plt.show()
```
