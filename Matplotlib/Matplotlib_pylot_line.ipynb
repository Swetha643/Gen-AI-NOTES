#**Linestyle**
You can use the keyword argument **linestyle**, or shorter **ls**, to change the style of the plotted line:

```python
import matplotlib.pyplot as plt

xpoints=[1,3,5]
ypoints=[2,5,8]

plt.plot(xpoints,ypoints,linestyle='--')
plt.show()
```

#Line Color
You can use the keyword argument **color** or the shorter **c** to set the color of the line:

```python
import matplotlib.pyplot as plt

xpoints=[1,3,5]
ypoints=[2,5,8]

plt.plot(xpoints,ypoints,linestyle='--',color='r')
plt.show()
```

#Line Width
You can use the keyword argument **linewidth** or the shorter **lw** to change the width of the line.

```python
import matplotlib.pyplot as plt

xpoints=[1,3,5]
ypoints=[2,5,8]

plt.plot(xpoints,ypoints,linestyle='--',color='hotpink',lw='5.5')
plt.show()
```

#Multiple Lines
You can plot as many lines as you like by simply adding more plt.plot() functions:

```python
import matplotlib.pyplot as plt

x1points=[1,2,3]
y1points=[2,5,8]

y2points=[4,5,6]

p1=plt.plot(x1points,y1points,marker='o',ls=':',c='g')
p2=plt.plot(y2points,marker="*",ls='--',c="hotpink")
plt.show()
```

#**Matplotlib Labels and Title**
#Create Labels for a Plot
With Pyplot, you can use the xlabel() and ylabel() functions to set a label for the x- and y-axis.

```python
import matplotlib.pyplot as plt

x1points=[1,2,3]
y1points=[2,5,8]

plt.plot(x1points,y1points,marker='o',ls=':',c='g')

plt.xlabel("Average plus")
plt.ylabel("calories burnage")

plt.show()
```

#Create a Title for a Plot
With Pyplot, you can use the **title()** function to set a title for the plot.

```python
import matplotlib.pyplot as plt

x1points=[0,1,2]
y1points=[2,4,6]

plt.plot(x1points,y1points,marker='o',ls=':',c='g')

plt.title("Sports watch Data")
plt.xlabel("Average plus")
plt.ylabel("calories burnage")

plt.show()
```

#Set Font Properties for Title and Labels using fontdict
You can use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels.

```python
import matplotlib.pyplot as plt

x1points=[0,1,2]
y1points=[2,4,6]

font1={"family":"serif","color":"red","size":20}
font2={"family":"serif","color":"blue","size":15}

plt.title("Sports watch Data",fontdict=font1)
plt.xlabel("Average plus",fontdict=font2)
plt.ylabel("calories burnage",fontdict=font2)

plt.plot(x1points,y1points,marker='o',ls=':',c='g')
plt.show()
```

#Position the Title
You can use the loc parameter in title() to position the title.

Legal values are: 'left', 'right', and 'center'. Default value is 'center'.

```python
import matplotlib.pyplot as plt

x1points=[0,1,2]
y1points=[2,4,6]

font1={"family":"serif","color":"red","size":20}
font2={"family":"serif","color":"blue","size":15}

plt.title("Sports watch Data",loc='Left',fontdict=font1)
plt.xlabel("Average plus",fontdict=font2)
plt.ylabel("calories burnage",fontdict=font2)

plt.plot(x1points,y1points,marker='o',ls=':',c='g')
plt.show()
```

#**Matplotlib Adding Grid Lines**
With Pyplot, you can use the grid() function to add grid lines to the plot.

```python
import matplotlib.pyplot as plt

# Create some data
x = [1, 2, 3]
y = [2, 4, 6]

# Plot the data
plt.plot(x, y)

plt.grid()
# Show the plot
plt.show()
```

#Specify Which Grid Lines to Display
You can use the axis parameter in the grid() function to specify which grid lines to display.

```python
import matplotlib.pyplot as plt

# Create some data
x = [1, 2, 3]
y = [2, 4, 6]

# Plot the data
plt.plot(x, y)

plt.grid(axis='y')
# Show the plot
plt.show()
```

#Set Line Properties for the Grid
You can also set the line properties of the grid, like this: grid(color = 'color', linestyle = 'linestyle', linewidth = number).

```python
import matplotlib.pyplot as plt

# Create some data
x = [1, 2, 3]
y = [2, 4, 6]

# Plot the data
plt.plot(x, y)

plt.grid(ls=':',c='r',lw='0.5')
# Show the plot
plt.show()
```
